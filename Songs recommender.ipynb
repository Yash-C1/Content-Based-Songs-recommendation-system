```python
import pandas as pd
import numpy as np
import ast
```

```python
songs = pd.read_csv("songs.csv", encoding='cp1252')
```

```python
songs.head()
```

```python
songs=songs.drop_duplicates()
```

```python
songs.drop(columns=['User-Rating'],axis=1,inplace=True)
```

```python
songs.head()
```

```python
songs = songs.rename(columns={"Singer/Artists":"Singers","Album/Movie":"Movie"})
```

```python
songs
```

```python
songs.isnull().sum()
```

```python
songs.dropna(inplace=True)
```

```python
songs.reset_index(drop=True,inplace=True)
```

```python
songs.shape
```

```python
songs
```

```python
for i in songs.columns:
    print(i)
    print(songs[i].unique())
    print("")
```

```python
def gap_fill(x):
    l=x.replace(" ","")
    return l
    
```

```python
#For singer/artists we need to remove all gaps between the 2 words
songs['Singers'] = songs['Singers'].apply(gap_fill)
```

```python
songs.head()
```

```python
# From genre, we will remove the word bollywood since it is present in all rows.
def bollywood_replace(x):
    l=x.replace("Bollywood","")
    return l

songs['Genre'] = songs['Genre'].apply(bollywood_replace)
```

```python
songs.head()
```

```python
#Now we need to seperate the individual singers
def singer_replace(x):
    l=x.replace(","," ")
    return l
songs['Singers'] = songs['Singers'].apply(singer_replace)
```

```python
songs.head()
```

```python
# Now we will create a tags column which will include the singers, genre and movie of the song
songs['tags'] = songs['Genre'] + " " + songs['Singers']
```

```python
songs.head()
```

```python
def lower(x):
    l= x.lower()
    return l
```

```python
songs['tags']=songs['tags'].apply(lower)
```

```python
songs['tags'][0]
```

# Data cleaning is done. Now we will perform vectorization on this data


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(songs['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
# Finding the cosine similarity based on distance between all vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(song):
    movie_list=[]
    song_index = songs[songs['Song-Name'] == song].index[0]
    print(song_index)
    distances = similarity[song_index]
    songs_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:15]
    
    for i in songs_list:
        movie_list.append(songs['Movie'].iloc[i[0]])
    
    #print(movie_list)
    for i in songs_list:
        if i[0] == song_index:
            continue
        print(songs['Song-Name'].iloc[i[0]])
```

```python
recommend('Tere Bin')
```
